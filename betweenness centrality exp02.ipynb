```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
```

```python
G.add_nodes_from([1,2,3,4,5,6,7,8,9])
```

```python
G.add_edge(1,2)
G.add_edge(2,4)
G.add_edge(3,4)
G.add_edge(4,5)
G.add_edge(4,6)
G.add_edge(6,7)
G.add_edge(7,8)
G.add_edge(8,9)
```

```python
nx.draw(G,with_labels=True)
```

```python
nx.betweenness_centrality(G)
```

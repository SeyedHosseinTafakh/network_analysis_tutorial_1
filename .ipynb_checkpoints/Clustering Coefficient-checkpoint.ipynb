```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
```

```python
G.add_nodes_from([1,2,3,4,5])
```

```python
G.add_edge(1,3)
G.add_edge(2,3)
G.add_edge(4,3)
G.add_edge(5,3)
G.add_edge(4,5)
nx.draw(G,with_labels=True)
```

```python
nx.algorithms.cluster.clustering(G)
```
